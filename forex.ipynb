```python
import pandas as pd

data = pd.read_csv("FOREXCOM_EURCAD, 5.csv")
data.head()
```

```python
data_mean = []
for info in data.iterrows():
    mean = (info[1]["open"] + info[1]["close"])/2
    data_mean.append(mean)
    
print(data_mean)
```

```python
from matplotlib import pyplot as plt
plt.plot(data_mean)
```

```python
selected_data = data_mean[-50:]
print(selected_data) 
```

```python
from scipy.stats.stats import pearsonr 
correlation = []
for i in range(0,len(data_mean)):
    print(i)
    if i < 1417:
        test_data = data_mean[i:i + 50]
        #print(len(test_data))
        #print(len(selected_data))
        correlation.append(pearsonr(selected_data, test_data))
```

```python
plt.plot(correlation)
```

```python
import numpy

corr = numpy.asarray(correlation)
print(corr)
```

```python
index = 0
for integer in corr:
    #print(integer)
    if integer[0] > 0.65:
#         print(index)
#         print(integer[0])
        plt.figure(index)
        plt.plot(data_mean[index:index+50])
        plt.title(str(index) +"_"+ str(integer[0]))
        plt.savefig("Result" + str(index) + ".pdf" )
        
    index += 1
    

plt.plot(data_mean[-50:])
plt.title("Original")
plt.savefig('Result.pdf')
```
